### Graphs

```python
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore", category=FutureWarning)
```

### Exercise 1

```python
df = pd.read_csv("Customers.csv")
df_customers = df.copy()
df_customers.dropna().drop_duplicates()

gender_counts = df_customers['gender'].value_counts()
fig = plt.figure()
ax = fig.add_axes([0, 0, 1.2, 1])
ax.bar(gender_counts.index, gender_counts.values, color=['b', 'tab:pink', 'g', 'c', 'm', 'y', 'k', 'r'])
ax.set_title('Gender Division')
ax.set_xlabel('Gender')
ax.set_ylabel('Count')
plt.show()
```

### Exercise 2

```python
df = pd.read_csv("Customers.csv")
df_customers = df.copy()
df_customers.dropna().drop_duplicates()

plt.figure(figsize=(12, 4))
sns.countplot(data=df_customers, x='nationallity', palette='Spectral', legend=False)
plt.tight_layout()
plt.show()
```
